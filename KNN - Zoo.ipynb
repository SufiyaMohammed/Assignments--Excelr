# Grid Search For Algorithm Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
zoo=pd.read_csv('Zoo.csv')
```

```python
zoo=zoo.drop(columns=['animal name'],axis=1)
array=zoo.values
x=array[:,0:16]
y=array[:,16]
```

```python
n_neighbors = np.array(range(1,80))
param_grid = dict(n_neighbors=n_neighbors)
```

```python
model = KNeighborsClassifier()
grid = GridSearchCV(estimator=model, param_grid=param_grid)
grid.fit(x, y)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
# choose k between 1 to 101
k_range = range(1, 80)
k_scores = []
# use iteration to caclulate different k in models, then return the average accuracy based on the cross validation
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, x, y, cv=5)
    k_scores.append(scores.mean())
# plot to see clearly
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```
